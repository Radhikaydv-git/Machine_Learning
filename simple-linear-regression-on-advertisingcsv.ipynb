```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('/kaggle/input/datasets/radydv/advertising-using-simple-lr/advertising.csv')
df
```

```python
df.isnull().sum()
```

```python
plt.scatter(df['TV'],df['Sales'])
plt.xlabel('Tv')
plt.ylabel('Sales')
```

```python
x = df.iloc[:,0:1]
x
```

```python
y = df.iloc[:,-1]
y
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y , test_size = 0.2 , random_state = 4)
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train , y_train)
```

```python
lr.predict(x_test)
```

```python
lr.predict(x_test.iloc[5:6])
```

```python
plt.scatter(df['TV'],df['Sales'])
plt.plot(x_train,lr.predict(x_train) , color = 'pink')
plt.xlabel('Tv')
plt.ylabel('Sales')
```

```python
m = lr.coef_ #m value 
```

```python
b = lr.intercept_ #b value
```
